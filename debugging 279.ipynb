```python
days = [1,2,3,4,5]
e = [3,5,4,6,2]
s = [8,9,6,10,7]
st = [1,3,4,6,7]
p = [5,6,8,14,24]
plt.plot([],[],color='k',label='e',linewidth=2)
plt.plot([],[],color='m',label='s',linewidth=2)
plt.plot([],[],color='c',label='st',linewidth=2)
plt.plot([],[],color='r',label='p',linewidth=2)
plt.stackplot(days,e,s,st,p)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np

def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)
t1 = np.arange(0.0,0.5,0.01)
t2 = np.arange(0.0,0.5,0.02)
plt.figure(1)

plt.subplot(211)
plt.plot(t1,f(t1),'b--')
plt.grid()
plt.show()

plt.subplot(212)
plt.plot(t2,np.cos(t2,np.pi*t2),'r--')
plt.show()
```

```python
import numpy as np

def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)
t1 = np.arange(1.0,5.0,0.1)
t2 = np.arange(1.0,5.0,0.2)
plt.figure(1)

plt.subplot(211)
plt.plot(t1,f(t1),'b--')
plt.grid()
plt.show()

plt.subplot(212)
plt.plot(t2,np.cos(t2,np.pi*t2),'r--')
plt.show()
```
