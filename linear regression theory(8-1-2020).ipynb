```python
plt.scatter(brain['Head Size(cm^3)'],brain['Brain Weight(grams)'])
plt.xlabel('Head Size(cm^3)')
plt.ylabel('Brain Weight(grams)')
plt.show()
```

```python
x = brain['Head Size(cm^3)'].values
y = brain['Brain Weight(grams)'].values
```

```python
x_mean = x.mean()
y_mean = y.mean()
```

```python
n = len(x)
num = 0
den = 0
for i in range(n):
    num +=(x[i]-x_mean) * (y[i]-y_mean)
    den +=(x[i]-x_mean)**2
    
M = (num)/(den)
print("slope M:",M)
```
