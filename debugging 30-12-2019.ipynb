```python
import math
l=[21,25,26,31,32,34,35,48,49,50,55]
def b(arr,l,r,x):
    if r>=1:
        mid=l+math.floor((r-1/2))
        if arr[mid]==x:
            return mid
        elif arr[mid]>x:
            return b(arr,1,mid-1,x)
        else:
            return b(arr,mid+1,r,x)
    else:
        return -1
l.sort(); 
arr=l;
q=31
b(arr,0,len(arr)-1,q)
```
